## Walkability dataset — columns we're using

> Column definitions and methodology sourced from the [EPA Smart Location Database Technical Documentation v3.0](https://www.epa.gov/system/files/documents/2023-10/epa_sld_3.0_technicaldocumentationuserguide_may2021_0.pdf)

| Column | What it is | Why we use it |
|---|---|---|
| `GEOID10` | 12-digit Census block group ID | Merge key — links walkability scores to the TIGER shapefile geometries |
| `CBSA` | Metro area code | Filter the national dataset down to San Diego only (code `41740`) |
| `CBSA_Name` | Metro area name | Sanity check that our filter worked correctly |
| `TotPop` | Total population per block group | Useful context for understanding how dense/populated each area is |
| `D1A` | Gross residential density (housing units/acre) | Upstream walkability input used by EPA to calculate `NatWalkInd`; can include placeholder values like `-99999` when data is missing |
| `D3B` | Street intersection density (intersections/sq mile) | Upstream walkability input used by EPA to calculate `NatWalkInd`; can include placeholder values like `-99999` when data is missing |
| `D4A` | Distance to nearest transit stop | Upstream walkability input used by EPA to calculate `NatWalkInd`; can include placeholder values like `-99999` when data is missing |
| `NatWalkInd` | National Walkability Index score (1–20) | EPA's precomputed composite score based on the D-variables, so we can keep just this score instead of carrying the raw D inputs in our beginner workflow |

> The D-variable columns above and all other raw SLD columns are dropped because `NatWalkInd` already incorporates them as a precomputed composite, making the ~110 individual inputs redundant for our project.

```python
cols_to_keep = [
    'GEOID10',       # block group ID — used to merge with TIGER shapefile
    'CBSA',          # filter to San Diego (code 41740)
    'CBSA_Name',     # sanity check that filter worked
    'TotPop',        # total population — useful context
    'NatWalkInd',    # national walkability index — our main scoring input (precomputed by EPA)
]

walkability_df = walkability_df_raw[cols_to_keep].copy()
walkability_df.head()
```

```python
# check nulls in the columns we kept and min/max to understand the range of values
print("Amount of data that isn't null and data types:")
print(walkability_df.info())

print("Summary statistics for numeric columns:")
print(walkability_df.describe())

print("Missing values in each column:")
print(walkability_df.isnull().sum())
```

```python
# understand GEOID10
# pandas read GEOID10 as numeric-looking values, so the raw display can include a trailing .0 and vary in length
raw_geoid10 = walkability_df['GEOID10'].astype(str)
print("Raw GEOID10 examples with lengths:")
for value in raw_geoid10.head():
    print(f"{value} | length: {len(str(value))}")

print("Unique raw GEOID10 lengths:", sorted(raw_geoid10.str.len().unique()))
for length in [14, 13, 3]:
    print(f"Examples of raw GEOID10 values with length {length}:")
    raw_examples = raw_geoid10[raw_geoid10.str.len() == length].head()
    if raw_examples.empty:
        print("None found")
    else:
        print(raw_examples.to_string(index=False))

# convert GEOID10 to clean text by removing the trailing .0 from numeric-looking values so the Census IDs can be compared consistently
def format_geoid10(value):
    if pd.isna(value):
        return value
    try:
        numeric_value = float(value)
        if numeric_value.is_integer():
            return str(int(numeric_value))
    except (TypeError, ValueError):
        pass
    return str(value)

geoid10_text = walkability_df['GEOID10'].map(format_geoid10)
print("Clean GEOID10 examples with lengths:")
for value in geoid10_text.head():
    print(f"{value} | length: {len(str(value))}")

# See if each cleaned GEOID10 has the same length (should be 12 for block groups)
geoid_lengths = geoid10_text.str.len()
unique_lengths = sorted(geoid_lengths.dropna().astype(int).unique())
print("Unique lengths of GEOID10 values:", unique_lengths)

# See examples of cleaned GEOID10 values for each length
for length in unique_lengths:
    print(f"Examples of GEOID10 values with length {length}:")
    print(geoid10_text[geoid_lengths == length].head().to_string(index=False))
```

```python
# Some GEOID10 values lose their leading zeros when read as numbers so they are length of 11 rather than 12, but they still represent the same block groups. We can confirm this by checking that the shorter values are all missing a leading zero and that adding a leading zero makes them match the longer values.

# Fill in the shorter GEOID10 values with a leading zero due to how pandas read them as numbers
geoid10_text = geoid10_text.apply(lambda x: '0' + x if pd.notna(x) and len(str(x)) == 11 else x)
print("GEOID10 examples after filling leading zeros:")
print(geoid10_text.head())
# Add the cleaned GEOID10 back to the DataFrame and drop the original raw GEOID10 column
walkability_df['GEOID10_clean'] = geoid10_text
walkability_df = walkability_df.drop(columns=['GEOID10'])
walkability_df.head()


```

```python
# Filter to only San Diego block groups using the cleaned GEOID10 values as we have nulls in the CBSA column
# Also GEOID10 is more reliable than CBSA for filtering to San Diego block groups since some of the CBSA values are null and we know that all San Diego block groups will have GEOID10 values that start with '060730'
# We also will have to merge with the TIGER shapefile using the cleaned GEOID10 values, so it's good to have this column ready for filtering and merging

walkability_df = walkability_df[walkability_df['GEOID10_clean'].str.startswith('060730')] # San Diego block groups start with '060730'
print(walkability_df.head())
print(f"We have {len(walkability_df)} block groups in San Diego after filtering by GEOID10.")

# Find unique CBSA Names
print("Unique CBSA Names in the filtered DataFrame:")
print(walkability_df['CBSA_Name'].unique())
```

## Merging cleaned walkability df 
We will merge the cleaned walkability DataFrame with the TIGER shapefile using `GEOID10_clean` as the shared key. This will attach polygon boundary geometries to each block group, giving each neighborhood a geographic shape that we can use to spatially join crime incidents and assign walkability scores to street edges in our routing algorithm.

```python
# load in the TIGER shapefile for San Diego block groups and check the GEOID10 values there to confirm they match the cleaned GEOID10 values in our walkability DataFrame
import geopandas as gpd
tiger_gdf = gpd.read_file('../data/raw/tl_2020_06_bg/tl_2020_06_bg.shp')
print("TIGER shapefile GEOID10 examples:")
print(display(tiger_gdf.head()))
```

```python
# check datatypes 
print("TIGER shapefile datatypes:")
print(tiger_gdf.dtypes)

# check for missing values 
print("Missing values in each TIGER shapefile column:")
print(tiger_gdf.isnull().sum().sort_values(ascending=False))
# check length of our GEOID values in tiger shapefile
tiger_gdf['GEOID10_length'] = tiger_gdf['GEOID'].astype(str).str.len() # column of length of each rows GEOID value

# unique lengths to see if they are the lengths the block groups are suppose to be
print("Unique lengths of GEOID10 values in TIGER shapefile:", sorted(tiger_gdf['GEOID10_length'].unique()))
print("Examples of GEOID10 values in TIGER shapefile with length 12:")
print(tiger_gdf[tiger_gdf['GEOID10_length'] == 12]['GEOID'].head())
```

```python
print(tiger_gdf.columns)

# keep important columns

tiger_gdf = tiger_gdf[['GEOID', 'geometry']]
print(display(tiger_gdf.head()))

```

```python
"""
Merge the walkability DataFrame with the Tiger shapefile GeoDataFrame on the 
cleaned GEOID10 values to create a GeoDataFrame that includes both the 
walkability data and the geometries for the block groups
"""

walkability_final_gdf = tiger_gdf.merge(walkability_df, left_on="GEOID", right_on="GEOID10_clean", how = "inner")
walkability_final_gdf = walkability_final_gdf.drop(columns=['GEOID10_clean'])
walkability_final_gdf.head()
```
